```python
import os
import pandas as pd
import numpy


archivos = os.listdir('./datastreams/')

archivos = [file for file in archivos if file.endswith('.csv') and 'provincia' in file]
archivos.append('ingresos_por_la_operacion_del_servicio_de_internet_fijo.csv')

diccio_nombres = {'df{}'.format(i+1): file for i, file in enumerate(archivos)}
```

```python
diccio_nombres
```

```python
provincias = pd.read_csv('./externaldata/provincias.csv')
id_provincias = provincias[['Provincia', 'id']].sort_values(by = 'Provincia', ascending = True)
```

## ARCHIVO --> acceso_a_internet_fijo_por_rangos_de_velocidad_de_bajada_y_provincia

```python
print(diccio_nombres['df3'])
df3 = pd.read_csv('./datastreams/' + diccio_nombres['df3'])
print(df3.shape)
df3.head()
```

```python
df3 = pd.merge(df3, id_provincias, how='left', on='Provincia')
```

```python
df3.drop(columns='Provincia', inplace=True)
```

```python
df3.rename( columns={'id':'id_provincia'}, inplace=True)
```

```python
df3
```

```python
df3.to_csv('./sql/' + diccio_nombres['df3'], index=False)
```
